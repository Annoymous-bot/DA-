```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
significance = 0.10

male_survived, total_male = (109, 577)

female_survived, total_female = (233, 314)
```

```python
survivied = np.array([male_survived, female_survived])

total = np.array([total_male, total_female])
```

```python
z_value, p_value = proportions_ztest(count=survivied, nobs=total,  alternative='two-sided')

print((abs(z_value),p_value))
```

p value is less than 0.05 then reject null hypotesis. Therefore significant difference between  proportions of survived females to that of survived
males.
