```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
headernames = ['sepal-length','sepal-width','petal-length','petal-width','Class']
```

```python
dataset = pd.read_csv(path,names=headernames)
print(type(dataset))
dataset.head(15)
```

1. Comparison between various species based on sepal length and width using respective
visualization


```python
plt.figure(figsize=(20,10))
plt.title("Comparison between various species based on sapel length and width")
sns.scatterplot(dataset['sepal-length'],dataset['sepal-width'],hue =dataset['Class'],s=50)
```

2.  Perform the Bi-variate Analysis via pairplot visualization


```python
sns.pairplot(dataset,hue='Class')
```

3. Justify Sepal Length and Sepal Width features are slightly correlated with each other

```python
dataset[['sepal-length','sepal-width']].corr()
```

```python
sns.heatmap(dataset[['sepal-length','sepal-width']].corr(),annot=True)
```

4. Visualize the categorical feature “Class” with all other four input variables

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot(y='petal-width', x= 'Class', data=dataset, ax=axes[0, 0])
sns.boxplot(y='petal-length', x= 'Class', data=dataset, ax=axes[0, 1])
sns.boxplot(y='sepal-width', x= 'Class', data=dataset, ax=axes[1, 0])
sns.boxplot(y='sepal-length', x= 'Class', data=dataset, ax=axes[1, 1])
```

5. Plot the density of the length and width in the class.

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.kdeplot(data=dataset,x='sepal-width',hue='Class',ax = axes[0,0])
sns.kdeplot(data=dataset,x='sepal-length',hue='Class',ax = axes[0,1])
sns.kdeplot(data=dataset,x='petal-width',hue='Class',ax = axes[1,0])
sns.kdeplot(data=dataset,x='petal-length',hue='Class',ax = axes[1,1])
```

6. Plot the probability density function(PDF) with each feature as a variable on X-axis and it’s histogram
and corresponding kernel density plot on Y-axis.


```python
sns.FacetGrid(dataset, hue="Class", height=5).map(sns.distplot, "sepal-length").add_legend()
sns.FacetGrid(dataset, hue="Class", height=5).map(sns.distplot, "sepal-width").add_legend()
sns.FacetGrid(dataset, hue="Class", height=5).map(sns.distplot, "petal-length").add_legend()
sns.FacetGrid(dataset, hue="Class", height=5).map(sns.distplot, "petal-width").add_legend()
```
